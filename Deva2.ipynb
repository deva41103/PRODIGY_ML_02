```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df['Annual Income (k$)'] = df['Annual Income (k$)'].astype(float)
df['Spending Score (1-100)'] = df['Spending Score (1-100)'].astype(float)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Annual Income (k$)', 'Spending Score (1-100)']])

n_clusters = 5

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(df_scaled)
labels = kmeans.labels_
df['Cluster'] = labels
print(kmeans.cluster_centers_)

```

```python
import matplotlib.pyplot as plt

plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-means clustering of customer purchase history')
plt.show()
```
